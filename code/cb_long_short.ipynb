```python
Zscore = lambda df:(df.sub(df.mean(axis = 1),axis = 0)).div(df.std(axis = 1),axis = 0)
非负化处理 = lambda df:df.add(df.min(axis = 1).abs(),axis = 0)

stock_turnover_rate = Data_dict['turnover']
stock_adj_high = Data_dict['Adj_High']
stock_adj_low = Data_dict['Adj_Low']
stock_adj_close = Data_dict['Adj_Close']
stock_adj_open = Data_dict['Adj_Open']

Turn20 = stock_turnover_rate.rolling(20).mean()
基准换手率 = stock_turnover_rate.rolling(40).mean().shift(20)
PctTurn = stock_turnover_rate / 基准换手率 - 1
PctTurn20 = PctTurn.rolling(20).mean()

振幅 = (stock_adj_high - stock_adj_low) / stock_adj_close.shift()
振幅换手率因子 = (PctTurn * 振幅).rolling(20).mean()

STR = stock_turnover_rate.rolling(20).std()
Vol20 = (stock_adj_close / stock_adj_close.shift() - 1).rolling(20).std()

基准换手率波动率 = stock_turnover_rate.rolling(40,min_periods = 1).std().shift(20)
SCR = STR / 基准换手率波动率.shift() - 1
SCR = SCR[np.isfinite(SCR)]

DRET = (stock_adj_close - stock_adj_open)/stock_adj_close.shift()
FAC = (stock_adj_high-stock_adj_low)/stock_adj_close.shift()
PLUS = (2 * stock_adj_close - stock_adj_high - stock_adj_low) / stock_adj_close.shift()

Turn20_mul_PLUS = 非负化处理(Zscore(Turn20)) * 非负化处理(Zscore(PLUS))
STR_mul_PLUS = 非负化处理(Zscore(STR)) * 非负化处理(Zscore(PLUS))

CCV = stock_adj_close.rolling(20).corr(stock_turnover_rate)
CDCV = stock_adj_close.diff().rolling(20).corr(stock_turnover_rate)
CCOV = (stock_adj_close-stock_adj_open).rolling(20).corr(stock_turnover_rate)
COV = COYCYV = (stock_adj_open-stock_adj_close.shift()).rolling(20).corr(stock_turnover_rate.shift())
RPV = Zscore(CCOV) - Zscore(COV)
Ret20 = (stock_adj_close / stock_adj_close.shift(20) - 1)

每日换手率变化率 = stock_turnover_rate / stock_turnover_rate.shift()-1
每日换手率变化率 = 每日换手率变化率[np.isfinite(每日换手率变化率)]
每日换手率变化率_MA20 = 每日换手率变化率.rolling(20).mean()
GTR = 每日换手率变化率.rolling(20).std()

换手因子_dict = {
    #'stock_turnover_rate':stock_turnover_rate,
    'Turn20':Turn20,
    'PctTurn':PctTurn,
    'PctTurn20':PctTurn20,
    '振幅':振幅,
    '振幅换手率因子':振幅换手率因子,
    'STR':STR,
    'Vol20':Vol20,
    'SCR':SCR,
    'DRET':DRET,
    'FAC':FAC,
    'PLUS':PLUS,
    
    'Turn20_mul_PLUS':Turn20_mul_PLUS,
    'STR_mul_PLUS':STR_mul_PLUS,

    'CCV':CCV,
    'CDCV':CDCV,
    'CCOV':CCOV,
    'COV':COV,
    'RPV':RPV,
    'Ret20':Ret20,

    '每日换手率变化率':每日换手率变化率,
    '每日换手率变化率_MA20':每日换手率变化率_MA20,
    'GTR':GTR,
 }
for k,v in tqdm.tqdm(换手因子_dict.items()):
    Data_dict[k] = v
```

```python
def max_drawdown(prices):
    # 計算累計的最大值
    cumulative_max = prices.cummax()
    # 計算回撤 (Drawdown)
    drawdown = (prices - cumulative_max) / cumulative_max
    # 計算最大回撤 (MDD)
    mdd = drawdown.min()
    return mdd
def show_bt(bt_ret):
    display(pd.concat({"CAGR(%)":bt_ret.mean() * 252,
            'Sharpe' : bt_ret.mean() / bt_ret.std()*252**0.5,
            'Calmar':bt_ret.calmar(),
            'MDD(%)' : bt_ret.max_drawdown()*100,
            '單利MDD(%)' : max_drawdown(bt_ret.cumsum().add(1))*100,
            '樣本勝率(%)' : bt_ret.apply(lambda X:((X.dropna()>0).sum() / X.dropna().shape[0])*100),
            '周勝率(%)' : bt_ret.apply(lambda X:((X.dropna().add(1).resample('W').prod().sub(1)>0).sum() / X.dropna().add(1).resample('W').prod().sub(1).dropna().shape[0])*100),
            '月勝率(%)' : bt_ret.apply(lambda X:((X.dropna().add(1).resample('ME').prod().sub(1)>0).sum() / X.dropna().add(1).resample('ME').prod().sub(1).shape[0])*100),
            '年勝率(%)' : bt_ret.apply(lambda X:((X.dropna().add(1).resample('YE').prod().sub(1)>0).sum() / X.dropna().add(1).resample('YE').prod().sub(1).shape[0])*100),
            'avg_win/avg_loss' : bt_ret.apply(lambda X:(X[X > 0].mean() / abs(X[X < 0].mean()))),
            'profit_factor' : bt_ret.profit_factor(),
            },axis = 1).round(2))
def backtest_factor(factor:pd.DataFrame,exp_ret:pd.DataFrame,rank_range_n:int = 10,start_date:str = '2020-01-01'):
    factor_rank = factor.rank(axis = 1,pct = True,method = 'first')

    IC_Se = factor.corrwith(exp_ret,axis=1,method='spearman').sort_index().loc[start_date:]
    print(f'IC_mean:{round(IC_Se.mean(),4)}')
    print(f'IC_IR:{round(IC_Se.mean()/IC_Se.std(),4)}')

    bt = pd.concat({f'{int(((_/rank_range_n)*100))}% ~ {int((_+1)/rank_range_n*100)}%':exp_ret[(factor_rank>_/rank_range_n) & (factor_rank<=(_+1)/rank_range_n)].mean(axis = 1) - exp_ret.mean(axis=1) for _ in tqdm.tqdm(range(rank_range_n))}, axis = 1).dropna(how = 'all')
    bt = bt.loc[start_date:]
    if (bt.iloc[:,-1] - bt.iloc[:,0]).add(1).prod() > 1:
        bt['LS_ret'] = bt.iloc[:,-1] - bt.iloc[:,0]
    else:
        bt['LS_ret'] = bt.iloc[:,0] - bt.iloc[:,-1]
    show_bt(bt)

    (bt.drop(columns='LS_ret').loc[start_date:].cagr()*100).iplot(kind = 'bar')
backtest_factor(factor,exp_ret,rank_range_n = 10,start_date = '2020-01-01')
```

```python
def calculate_return_series(
    ret,
    Low_N,
    AUM,
    adj_close,
    exp_ret,
    bench_exp_ret,
    y_pred,
    days_for_trend= 5 
):

    # =========== (0) 大盤 / 產業走空判斷 ===========
    market_trend = bench_exp_ret.shift(2).rolling(window=days_for_trend).mean()
    market_signal = (market_trend < 0).astype(int)
    ret = ret.loc['2020':].fillna(0)

    # (B) 每日找出排名最低的Low_N檔 => short_signal = -1
    ranking_asc = ret.rank(axis=1, ascending=True, method='first')
    short_signal = (ranking_asc <= Low_N).astype(int) * -1

    y_pred = y_pred.shift(1)
    model_signal = pd.DataFrame(
        index=short_signal.index,
        columns=short_signal.columns,
        data=np.tile((y_pred == 0).values.reshape(-1,1), (1, short_signal.shape[1]))
    )
    short_signal = short_signal * model_signal

    # (D) 大盤/產業空頭濾網：若當天大盤不空 => 不做空
    market_signal_df = pd.DataFrame(
        index=short_signal.index,
        columns=short_signal.columns,
        data=np.tile(market_signal.reindex(short_signal.index).values.reshape(-1,1),
                     (1, short_signal.shape[1]))
    )
    short_signal = short_signal * market_signal_df

    # (E) 每日計算持倉 (weightings)
    W = short_signal / Low_N
    weighting = ((W * AUM) / (1000 * adj_close)) * (1000 * adj_close) / AUM
    weighting.replace({0: np.nan}, inplace=True)

    # (F) 計算每日組合損益 (short_return)
    def calculate_return(wgt):
        delta_weighting = wgt.diff()
        day_buy_fee  = delta_weighting[delta_weighting > 0] * (0.001425 * 0.3 + 0.003)
        day_sell_fee = delta_weighting[delta_weighting < 0].abs() * (0.001425 * 0.3)

        return (wgt * exp_ret).sum(axis=1) - day_buy_fee.fillna(0).sum(axis=1) - day_sell_fee.fillna(0).sum(axis=1)

    short_return = calculate_return(weighting)

    bt_ret = pd.concat({
        'short_return': short_return,
        'benchmark': bench_exp_ret.reindex(exp_ret.index)
    }, axis=1).loc['2020':]

    return bt_ret, weighting
```
